# Best Composition Mean/Std Graph

```python
importantFiles = ['~/activemeltheas/data/Simulated_TestingFarkas_Ni50_Fe50_MLIBest100Pico.csv','~/activemeltheas/data/Simulated_TestingFarkas_Ni40_Fe50_Co10_MLIBest100Pico.csv','~/activemeltheas/data/Simulated_TestingFarkas_Ni30_Fe50_Co20_MLIBest100Pico.csv',
                 '~/activemeltheas/data/Simulated_TestingFarkas_Ni20_Fe50_Co30_MLIBest100Pico.csv','~/activemeltheas/data/Simulated_TestingFarkas_Ni10_Fe50_Co40_MLIBest100Pico.csv','~/activemeltheas/data/Simulated_TestingFarkas_Fe50_Co50_MLIBest100Pico.csv']

compositional_list = []
compositional_list_mean = []
compositional_list_max = []
compositional_list_min = []

for comps in importantFiles:
    data = pd.read_csv(comps)
    compositional_list.append(data['Simulated Tm'])
    compositional_list_mean.append(st.mean(data['Simulated Tm']))
    compositional_list_max.append(max(data['Simulated Tm'])-min(data['Simulated Tm']))
    compositional_list_min.append(st.stdev(data['Simulated Tm']))
```

```python
fig = plt.figure(figsize = (10,10))

plt.errorbar([0,10,20,30,40,50],compositional_list_mean,fmt='o',markersize=20,yerr=compositional_list_min,capsize =10)
plt.xticks(fontsize=26)
plt.yticks(fontsize=26)
plt.grid()
plt.legend(fontsize=22)
plt.ylabel('Melting Temperature (K)', fontsize=26)
plt.xlabel('Atomic % Cobalt', fontsize=26)
```
