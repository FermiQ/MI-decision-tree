## Overview

In this work, we pair AL with MD simulations to find multiple principal component alloys (MPCAs) with high predicted melting temperature. The following notebook outlines our workflow and our results, which are available as a publication as <b>[XXXXX]</b>. <br> Users can modify the notebooks to explore different models, try new ideas and adapt them for their own problems.

In this example, we aim to solve a materials design problem to efficiently screen a high-dimensional space of compositions to find high-performing alloy systems. Using a predictive model based on Random Forests (RF) with uncertainty estimates based on [Ling et al.](https://link.springer.com/article/10.1007/s40192-017-0098-z), we iteratively explore the compositional space with different acquisition functions to query the next point.
